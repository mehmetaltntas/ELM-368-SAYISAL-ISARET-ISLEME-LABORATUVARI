# SORU-1 

```python
import numpy as np

# a ve b bileşenlerini tanımlama
a = 3 - 4j
b = 1 + 2j

# C değerini hesaplama
C = a + b

# İstenen değerler için yapılan atamalar
real_part = C.real
imaginary_part = C.imag
magnitude = abs(C)
phase = np.angle(C)

# Sonuçları print fonksiyonu ile ekrana yazdırma
print("I. Gerçek Bileşen :", real_part)
print("II. Sanal Bileşen :", imaginary_part)
print("III. Genlik :", magnitude)
print("IV. Faz (radyan):", phase)
```

# SORU-2

```python
import numpy as np
import matplotlib.pyplot as plt

# Periyotları belirleme
T1 = 5  # x1[n] için periyot x1[n] = 2sin((2π/5)n - π/5)
T2 = 10  # x2[n] için periyot x2[n] = sin((3π/5)n - 2π/5)

# Zaman dizileri (2 tam periyot olacak şekilde ayarlama)
n1 = np.arange(0, 2 * T1, 1)  # x1 için 2 tam periyot
n2 = np.arange(0, 2 * T2, 1)  # x2 için 2 tam periyot

# İşaretleri hesaplama
x1 = 2 * np.sin((2 * np.pi / 5) * n1 - np.pi / 5)
x2 = np.sin((3 * np.pi / 5) * n2 - 2 * np.pi / 5)

# Grafikleri çizdirme
plt.figure(figsize=(10, 6))

# Birinci işaret (x1)
plt.subplot(2, 1, 1)
plt.stem(n1, x1, linefmt='b-', markerfmt='bo', basefmt='k')
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.title("x1[n] = 2sin((2π/5)n - π/5) - 2 Tam Periyot")
plt.axvline(x=T1, color='r', linestyle='--', label="T = 5")
plt.axvline(x=2*T1-1, color='r', linestyle='--')
plt.legend()
plt.grid(True)

# İkinci işaret (x2)
plt.subplot(2, 1, 2)
plt.stem(n2, x2, linefmt='g-', markerfmt='go', basefmt='k')
plt.xlabel("n")
plt.ylabel("x2[n]")
plt.title("x2[n] = sin((3π/5)n - 2π/5) - 2 Tam Periyot")
plt.axvline(x=T2, color='r', linestyle='--', label="T = 10")
plt.axvline(x=2*T2-1, color='r', linestyle='--')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
t1 = np.arange(0, 16, 1)


# x1[n] = e^(π/5)
x1 = np.exp(np.pi / 5) * np.ones_like(t1)
plt.figure(figsize=(6, 4))
plt.stem(t1, x1, markerfmt='bo', linefmt='b-', basefmt='k')
plt.xlabel("n")
plt.ylabel("x1[n]")
plt.title("x1[n] = e^(π/5)")
plt.grid()
plt.show()

```

##  $x_1[n]$ sinyali n'den bağımsız bir sinyal olduğundan n'in değişen değerleri için sabit bir değer almaktadır.

# 3.3

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
n = np.arange(0, 16)  # n = 0,1,...,15

# x₃[n] = e^(jπn/5)
x3 = np.exp(1j * np.pi * n / 5)

# Gerçek ve sanal kısımları
real_part = np.real(x3)
imag_part = np.imag(x3)

# Grafiği çizdirme
plt.figure(figsize=(12, 6))

# Gerçek kısım
plt.subplot(2, 1, 1)
plt.stem(n, real_part, 'b', markerfmt='bo', basefmt="k")
plt.grid(True)
plt.title("x_3[n] = e^(jπn/5) Gerçek Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

# Sanal kısım
plt.subplot(2, 1, 2)
plt.stem(n, imag_part, 'r', markerfmt='ro', basefmt="k")
plt.grid(True)
plt.title("x_3[n] = e^(jπn/5) Sanal Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

plt.tight_layout()
plt.show()
```

#  $x_3[n]$ kompleks üstel bir fonksiyondur. $ \omega = 2 \pi f = \pi / 5 \Longrightarrow f = \frac{1}{10} \Longrightarrow T = 10 $ periyodik bir bir sinyaldir.

# 3.4

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
n = np.arange(0, 16)  # n = 0,1,...,15

# x₄[n] = cos(π/5) + j sin(π/5)
# Bu işaret, tüm n değerleri için sabit bir kompleks sayıdır
x4 = np.cos(np.pi/5) + 1j * np.sin(np.pi/5)
x4_array = np.full_like(n, x4, dtype=complex)

# Gerçek ve sanal kısımları
real_part = np.real(x4_array)
imag_part = np.imag(x4_array)

# Grafiği çizdirme
plt.figure(figsize=(12, 6))

# Gerçek kısım
plt.subplot(2, 1, 1)
plt.stem(n, real_part, 'b', markerfmt='bo', basefmt="k")
plt.grid(True)
plt.title("x_4[n] = cos(π/5) + j sin(π/5) Gerçek Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

# Sanal kısım
plt.subplot(2, 1, 2)
plt.stem(n, imag_part, 'r', markerfmt='ro', basefmt="k")
plt.grid(True)
plt.title("x_4[n] = cos(π/5) + j sin(π/5) Sanal Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

plt.tight_layout()
plt.show()
```

## $x_4[n]$ işaretinin grafiği ile $x_2[n]$ işaretinin grafiği aynı olduğu gözlemlenmektedir. Bu durum euler formülü ile açıklanabilir: $ e^{j* \alpha } = cos( \alpha ) + j*sin( \alpha ) $

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
n = np.arange(0, 16)  # n = 0,1,...,15

# x₅[n] = cos(πn/5) + j sin(πn/5)
x5 = np.cos(np.pi * n / 5) + 1j * np.sin(np.pi * n / 5)

# Gerçek ve sanal kısımları
real_part = np.real(x5)
imag_part = np.imag(x5)

# Grafiği çizdirme
plt.figure(figsize=(12, 6))

# Gerçek kısım
plt.subplot(2, 1, 1)
plt.stem(n, real_part, 'b', markerfmt='bo', basefmt="k")
plt.grid(True)
plt.title("x_5[n] = cos(πn/5) + j sin(πn/5) Gerçek Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

# Sanal kısım
plt.subplot(2, 1, 2)
plt.stem(n, imag_part, 'r', markerfmt='ro', basefmt="k")
plt.grid(True)
plt.title("x_5[n] = cos(πn/5) + j sin(πn/5) Sanal Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
n = np.arange(-10, 5)  # n = -10,-9,...,3,4

# x₆[n] = cos(πn/5) + j sin(πn/5)
x6 = np.cos(np.pi * n / 5) + 1j * np.sin(np.pi * n / 5)

# Gerçek ve sanal kısımları
real_part = np.real(x6)
imag_part = np.imag(x6)

# Grafiği çizdirme
plt.figure(figsize=(12, 6))

# Gerçek kısım
plt.subplot(2, 1, 1)
plt.stem(n, real_part, 'b', markerfmt='bo', basefmt="k")
plt.grid(True)
plt.title("x_6[n] = cos(πn/5) + j sin(πn/5) Gerçek Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

# Sanal kısım
plt.subplot(2, 1, 2)
plt.stem(n, imag_part, 'r', markerfmt='ro', basefmt="k")
plt.grid(True)
plt.title("x_6[n] = cos(πn/5) + j sin(πn/5) Sanal Kısım")
plt.xlabel("n")
plt.ylabel("Genlik")

plt.tight_layout()
plt.show()
```

## $x_5[n]$ işaretinin grafiği ile $x_6[n]$ işaretinin grafiği aynı olduğu gözlemlenmektedir. Bu gözlem periyodik bir sinyal için oldukça normaldir. T = 10 periyodik bir sinyalin 10k (k = ...-3,-2,-1,0,1,2,3...) birim ötelenmesi sinyalde bir değişime sebep olmayacaktır. $x_6[n]$ sinyali, $x_5[n]$ sinyalinin 10 birim sola ötelenmeiş hali olduğundan sinyaller eşdeğer grafiklere sahiptir.

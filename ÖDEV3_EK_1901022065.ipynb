# SORU 2 - EK

```python
import numpy as np
import matplotlib.pyplot as plt

# (a) verilen bir periyotluk dizi
x = np.array([0, 0.707106, -1, 0.707106, 0, -0.707106, 1, -0.707106])
N = len(x)

# 8-noktalı DFT
X = np.fft.fft(x, N)
k = np.arange(N)
omega_k = 2*np.pi * k / N

# genlik / faz
X_abs = np.abs(X)
X_ph  = np.angle(X)

# çizimler
plt.figure(figsize=(8,3.5))

plt.subplot(121)
plt.stem(omega_k/np.pi, X_abs,
         linefmt='b-', markerfmt='bo', basefmt='k-')
plt.title(r'$|X[k]|$')
plt.xlabel(r'$\omega_k/\pi$')
plt.grid(True)

plt.subplot(122)
plt.stem(omega_k/np.pi, X_ph,
         linefmt='r-', markerfmt='ro', basefmt='k-')
plt.title(r'$\angle X[k]$')
plt.xlabel(r'$\omega_k/\pi$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# SORU 3 - EK

## $x[n] $ = cos$(\frac {n\pi} {3})$ işareti, $x[n] $ = $δ[n]$ - $δ[n-1]$ sistemine girince ...

## $ H(e^{jw})  = \mathtt{1 - e^{-jw}}$  formülünden 
## $ H(e^{j(\frac {\pi} {3})})  = \mathtt{1 - e^{-j(\frac {\pi} {3})}}$ = 0.5 + $j0.8660$ ⟹ $∣H∣$ = 1 , $∠H$ = $(\frac {\pi} {3})$

## Bu nedenle çıkış

## $y[n] $ = $∣H(e^{j\frac {\pi} {3}})∣$ = cos$(\frac {n\pi} {3} + ∠H(e^{j(\frac {\pi} {3})}))$ = cos$(\frac {n\pi} {3} + \frac {\pi} {3})$ 

```python
import numpy as np
import matplotlib.pyplot as plt

# (1) H değerleri
omega0 = np.pi/3
H      = 1 - np.exp(-1j*omega0)


# (2) y[n] hesapla ve çiz
n = np.arange(0, 11)
y = np.cos(omega0*n + np.angle(H))

plt.stem(n, y, linefmt='b-', markerfmt='bo', basefmt='k-')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title(r'$y[n]=\cos\!\left(\frac{\pi}{3}n + \frac{\pi}{3}\right)$')

plt.grid(True)
plt.show()

print('|H(e^{jπ/3})| =', np.abs(H))
print('∠H(e^{jπ/3}) =', np.angle(H), 'rad')
```

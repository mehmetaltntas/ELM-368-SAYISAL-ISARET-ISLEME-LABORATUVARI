# SORU 1-EK

# a) Bu sistemin dürtü cevabını el yordamıyla bulunuz

### Sistemin fark denklemi:

## $y[n]−1.2y[n−1]=x[n]−x[n−4]$ olarak verilmiştir

### Dürtü cevabını bulmak için $x[n]=δ[n]$  koyduğumuzda denklem;

## $h[n]−1.2h[n−1]=δ[n]−δ[n−4]$ halini alır.

### İlk olarak n<0 için h[n]=0 alınır. 

## n = 0: h[0] - 1.2h[-1] = δ[0] - δ[-4] → h[0] = 1 , 
## n = 1: h[1] - 1.2h[0] = 0 → h[1] = 1.2 , 
## n = 2: h[2] = 1.2h[1] = $(1.2)^2$ → 1.44 , 
## n = 3: h[3] = 1.2h[2] = $(1.2)^3$ → 1.728 , 
## n = 4: h[4] = 1.2h[3] = $(1.2)^4$ - 1 → 1.0736 ,

## Genel olarak,  n≥4 için: $(1.2)^n$ - (1.2)^(n-4)

### Dolayısıyla, impuls cevabı şu şekilde ifade edilebilir: 

## h[n] = $(1.2)^n u[n]$  - (1.2)^(n-4)u[n-4]

# b) Bu sistem kararlı mıdır? Neden?

### Bir LTI sistemin kararlı olabilmesi için impuls cevabının mutlak toplamı $\left(\sum \:_{n=-\infty \:}^{\infty \:}\:∣h\left[n\right]∣\:\right) $ sonlu olmalıdır. Burada $(1.2)^n u[n]$  - $(1.2)^(n-4)u[n-4]$ ifadesinde, $(1.2)^n u[n]$ $∣(1.2)^n∣$ olarak düşünüldüğünde geometrik bir dizi oluşturur ve 1.2 > 1 olduğundan bu dizi diverjan (sonsuz büyür). Bu durum, impuls cevabının mutlak toplamının sonsuza gitmesine neden olur. Sonuç olarak, sistemin dürtü cevabı mutlak toplamı sonsuz olduğundan sistem kararsızdır.

# SORU 2-EK

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve

# Sistem-1'in impuls cevabı: h1[n] = 1/3*(δ[n+1] + δ[n] + δ[n-1])
# Dizisel gösterimde; δ[n+1] -> sol kayma, δ[n-1] -> sağ kayma.
# h1' i merkezleştirilmiş 3 noktalı dizi olarak kullanıyoruz:
h1 = np.array([1/3, 1/3, 1/3])  # Bu dizi, n = -1, 0, 1 indekslerini temsil eder.

# ------------------- (a) -------------------
# x1[n] = cos((pi/5)*n)*(u[n-10]-u[n-20])
n1 = np.arange(0, 30)  # n aralığını 0-29 olarak belirledik.
# u[n-10]-u[n-20]: n>=10 -> 1, n>=20 -> 1 olduğundan fark 1, aksi halde 0.
x1 = np.cos((np.pi/5)*n1) * (((n1 >= 10).astype(int)) - ((n1 >= 20).astype(int)))

# Çıkışı hesaplamak için konvolüsyon
y1 = convolve(x1, h1, mode='full')
n_y1 = np.arange(0, len(y1))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(n1, x1, basefmt="k")
plt.title("Giriş: x₁[n]")
plt.xlabel("n")
plt.ylabel("x₁[n]")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(n_y1, y1, linefmt='g-', markerfmt='go',  basefmt="k")
plt.title("Çıkış: y₁[n]")
plt.xlabel("n")
plt.ylabel("y₁[n]")
plt.grid(True)
plt.tight_layout()
plt.show()

# ------------------- (b) -------------------
# x2[n] = (-0.3)^n * (u[n-3]-u[n-5])
n2 = np.arange(0, 15)  # n aralığını 0-14 olarak belirledik.
x2 = ((-0.3)**n2) * (((n2 >= 3).astype(int)) - ((n2 >= 5).astype(int)))

# Konvolüsyon ile çıkış hesaplanıyor.
y2 = convolve(x2, h1, mode='full')
n_y2 = np.arange(0, len(y2))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(n2, x2, basefmt="k")
plt.title("Giriş: x₂[n]")
plt.xlabel("n")
plt.ylabel("x₂[n]")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(n_y2, y2, linefmt='g-', markerfmt='go', basefmt="k")
plt.title("Çıkış: y₂[n]")
plt.xlabel("n")
plt.ylabel("y₂[n]")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# SORU 3-EK

# a) Bu sistemin dürtü cevabını el yordamıyla bulunuz.

### Sistemin dürtü cevabını (impulse response) elde etmek için girişe $x[n] = δ[n]$ verilir. Böylece denklem 

# $h[n]−0.8h[n−1] = δ[n]−δ[n−4]$ haline gelir. 

## İlk olarak $n<0$ için h[n]=0 varsayalim. 

## n = 0: h[0] - 0.8h[-1] = δ[0] - δ[-4] → h[0] = 1 , 
## n = 1: h[1] - 0.8h[0] = 0 → h[1] = 0.8, 
## n = 2: h[2] = 0.8h[1] = 0 → h[2] = 0.8h[1] = $(0.8)^2$ = 0.64  , 
## n = 3: h[3] = 0.8h[2] = 0 → h[3] = 0.8h[2] = $(0.8)^3$ = 0.512 , 
## n = 4: h[4] = 0.8h[3] = 0 - 1 → h[4] = 0.8h[3] -1 = $(0.8)^4$ - 1 ,

## Genel olarak,  $0 ≤ n < 4$ için: h[n] =  $(0.8)^n$

### ve 𝑛≥4  için, dürtü girişinin etkisi (–1) eklenerek

## h[n] = $(0.8)^n$ - (0.8)^(n-4)

### Dolayısıyla, impuls cevabı şu şekilde ifade edilebilir:

# h[n] = $(0.8)^n$ u[n]- (0.8)^(n-4) u[n-4]

# b) Bu sistem kararlı mıdır? Neden?

### Bir LTI sistemin kararlı olabilmesi için dürtü cevabının mutlak toplamı sonlu olmalıdır. Burada ℎ[𝑛] iki kısımdan oluşmaktadır:

## $(0.8)^n$ u[n]: 0.8 < 1 olduğundan bu terim üstel olarak azalarak sonlu bir toplam verir.
## (0.8)^(n-4) u[n-4]: Bu terim de aynı şekilde üstel olarak azalmaktadır.

## Sonuç olarak, sistemin dürtü cevabı mutlak toplamı sonlu olduğundan sistem kararlıdır.

# c)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, convolve

# Fark denklemindeki katsayılar:
b = [1, 0, 0, 0, -1]   # x[n] - x[n-4]
a = [1, -0.8]          # y[n] - 0.8*y[n-1]

# n = 0,1,...,10 aralığında x[n]=u[n-3] tanımlanıyor
n = np.arange(0, 11)
x = np.where(n >= 3, 1, 0)

# lfilter() yöntemi ile çıkış hesaplanıyor:
y_lfilter = lfilter(b, a, x)

# Sistemin dürtü cevabını (h[n]) elde etmek için delta sinyali kullanıyoruz:
delta = np.zeros(20)
delta[0] = 1
h = lfilter(b, a, delta)

# convolve() yöntemi: 'same' modu kullanılarak girişle aynı boyutta çıkış elde ediliyor.
y_conv = convolve(x, h, mode='same')

# Grafik çizimleri:
plt.figure(figsize=(12, 10))

plt.subplot(3,1,1)
plt.stem(n, x, basefmt="k")
plt.title("Giriş: x[n] = u[n-3]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(n, y_lfilter, linefmt='g-', markerfmt='go', basefmt="k")
plt.title("Çıkış (lfilter ile)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)

plt.subplot(3,1,3)
# y_conv dizisinin uzunluğu, 'same' modu nedeniyle girişle aynı boyutta
n_conv = np.arange(0, len(y_conv))
plt.stem(n_conv, y_conv, linefmt='r-', markerfmt='ro', basefmt="k")
plt.title("Çıkış (convolve ile, mode='same')")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Sonuçların Karşılaştırılması

### lfilter(): Fark denklemi kausal olarak uygulanır; yani n<0 için başlangıç değerleri sıfır kabul edilir ve sistemin gerçek zaman gecikmeleri korunur.
### convolve() (mode='same'): Çıkış sinyalini girişle aynı boyutta döndürmek için konvolüsyon sonucu merkezlenir. Bu durumda, sistemin asimetrik (gecikmeli) dürtü cevabındaki kayma tam olarak yansıtılamayabilir.

### İki yöntem arasında farklılığın temel nedeni yukarıdaki indeks hizalaması (time alignment) farkıdır. lfilter() nedensel yapıyı korurken, convolve()’daki 'same' modu zaman merkezlemesi yaptığından, özellikle sistemin dürtü cevabının simetrik olmaması durumunda sonuçlarda ofset gözlenebilir.

# SORU-1

```python
import numpy as np

# a ve b bileşenlerini tanımlama
a = 3 - 4j
b = 1 + 2j

# D değerini hesaplama
D = a / b

# İstenen değerler için yapılan atamalar
real_part = D.real
imaginary_part = D.imag
magnitude = abs(D)
phase = np.angle(D)

# Sonuçları print fonksiyonu ile ekrana yazdırma
print("I. Gerçek Bileşen :", real_part)
print("II. Sanal Bileşen :", imaginary_part)
print("III. Genlik :", magnitude)
print("IV. Faz (radyan):", phase)
```

# SORU-2

```python
import numpy as np
import matplotlib.pyplot as plt

# T1  = 2pi/(2pi/5) , T1 = 5
# T2  = 2pi/(3pi/5) , T2 = 10/3
# EKOK(5, 10/3) = 10 , T = 10
# Iki periyot çizdirecegimizden 2T = 20 uzunlugunda çizdireceğiz

# Aralık belirleme (0'dan 19'a kadar)
n = np.arange(0, 20, 1)

# x[n] fonksiyonunu hesaplama
x_n = 2 * np.sin((2 * np.pi / 5) * n - np.pi / 5) + np.sin((3 * np.pi / 5) * n - 2 * np.pi / 5)

# Grafiği çizdirme
plt.figure(figsize=(10, 4))
plt.stem(n, x_n)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("x[n] = 2sin((2π/5)n - π/5) + sin((3π/5)n - 2π/5)")
plt.grid(True)
```

# SORU-3

```python
import numpy as np
import matplotlib.pyplot as plt

# İndis vektörü
n = np.arange(0, 16, 1)

# Sabit üstel işaret
x_2 = np.exp(1j * np.pi / 5) * np.ones_like(n)

# Grafik çizimi
plt.figure(figsize=(10, 6))

# Gerçek kısım
plt.subplot(2, 1, 1)
plt.stem(n, np.real(x_2), markerfmt='bo', linefmt='b-', basefmt='k')
plt.xlabel("n")
plt.ylabel("Gerçek Kısım")
plt.title("Gerçek Kısım: Re{x_2[n]}")
plt.grid(True)

# Sanal kısım
plt.subplot(2, 1, 2)
plt.stem(n, np.imag(x_2), markerfmt='ro', linefmt='r-', basefmt='k')
plt.xlabel("n")
plt.ylabel("Sanal Kısım")
plt.title("Sanal Kısım: Im{x_2[n]}")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## $e^{j*\frac{\pi}{5}}$ifadesi, Euler formülü kullanılarak  $cos(\frac{\pi}{5}) + j*sin(\frac{\pi}{5})$ şeklinde yazılabilir. Bu durum, tanımlanan $x_2[n]$ işaretinin n indisinden bağımsız sabit bir kompleks sayı olduğunu gösterir. Diğer bir deyişle, n değişse bile $x_2[n]$ her zaman aynı değeri alır.

# Kompleks bir sayı olduğu için, bu işaret iki bileşenden oluşur:

# Gerçek kısım: sabit $cos(\frac{\pi}{5})$  değeri
# Sanal kısım: sabit $sin(\frac{\pi}{5})$ değeri

# Bu nedenle, farklı n değerleri için işaretin grafiği çizildiğinde, hem gerçek hem de sanal kısım için tüm n değerlerinde aynı genlikte sabit bir düz çizgi elde edilir. Bu, işaretin zaman (veya indis n) ile değişmediğini, sabit bir kompleks değer olduğunu gösterir.
